3.8 Addiing Coloumn for"Total Score" and "Average"  it is asically fetuter enginnering

```python
df['total_score']=df['math_score']+ df['reading_score']+df['writing_score']  #create new features that might be more informative or useful for a specific task or analysis. 
# By combining multiple columns into one, you might be capturing a higher-level abstraction or creating a new feature that encapsulates the information from those three columns in a more concise or meaningful way. 
# This can often help improve the performance of machine learning models or simplify downstream analysis. So yes, you're right to think of it as feature engineering.
df['average']=df['total_score']/3  #  now these total_score and average will be dependent features 
df.head()
```

```python
df[df['reading_score']==100]['average'].count()
```

```python
reading_full=df[df['reading_score'] == 100]['average'].count()  # forst we caclautle ereading score===10 than take out average of tehm adn than take count
writing_full=df[df['writing_score'] == 100]['average'].count()
math_full=df[df['math_score'] == 100]['average'].count()

print(f'number of students with full marks in maths: {math_full}')
print(f'number of students with full marks in writing: {writing_full}')
print(f'number of students with full marks in reading: {reading_full}')
```

```python
reading_less_20=df[df['reading_score'] <= 20]['average'].count()
writing_less_20=df[df['writing_score'] <= 20]['average'].count()
math_less_20=df[df['math_score'] <= 20]['average'].count()

print(f'number of student with less than 20 marks in maths: {math_less_20}')
print(f"number of students with less than 20 marks in writing:{writing_less_20}")
print(f"number of students with less than 20 marks in reading:{reading_less_20}")
```

Insights:

From above values we get students have performed the worst in Maths

Best performance is in reading section

4.1.1 Histogram & KDE

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='average',bins=30,kde=True,color='g')  # tiss i the simple graph of average
plt.subplot(122)
sns.histplot(data=df,x='average',kde=True,hue='gender') # thiss i the graph correpsonding to male and female 
# in this grpah we observed that averageid increaing along witht e female value
plt.show()
```

```python
# above data with respoect to total score
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='total_score',bins=30,kde=True,color='g')  # tiss i the simple graph of total_score
plt.subplot(122)
sns.histplot(data=df,x='total_score',kde=True,hue='gender') # thiss i the graph correpsonding to male and female 
# kde=True: Adds a kernel density estimate curve to the plot.
# in this grpah we observed that averageid increaing along witht e female value
plt.show()

```
